```python
daily_trader_perf = trade.groupby(['Account', 'date']).agg(
    total_trades=('Size USD', 'count'),
    total_volume_usd=('Size USD', 'sum'),
    total_pnl=('Closed PnL', 'sum'),
    avg_pnl=('Closed PnL', 'mean'),
    avg_trade_size=('Size USD', 'mean')
).reset_index()
```

```python
fg['date'] = pd.to_datetime(fg['date']).dt.date
```

```python
merged_df = pd.merge(daily_trader_perf, fg[['date', 'value', 'classification']], on='date', how='left')
```

```python
merged_df['avg_pnl_per_trade']=merged_df['total_pnl']/merged_df['total_trades']
```

```python
merged_df[['classification', 'total_trades', 'total_pnl', 'avg_pnl_per_trade']]
```

```python
# Aggregate total trades and total PnL per sentiment
sentiment_summary = merged_df.groupby('classification').agg({
    'total_trades': 'sum',
    'total_pnl': 'sum'
}).reset_index()


```

```python
# Compute average PnL per trade
sentiment_summary['avg_pnl_per_trade'] = sentiment_summary['total_pnl'] / sentiment_summary['total_trades']
```

```python
sentiment_summary
```

The most profitable trades per unit happen in Extreme Greed and Fear markets. Extreme Fear generates lots of activity but smaller trade profits per trade.

```python
# Step 1: Compute PnL per USD for each row
merged_df['pnl_per_usd'] = merged_df['total_pnl'] / merged_df['total_volume_usd']

# Step 2: Aggregate by sentiment
pnl_efficiency_summary = merged_df.groupby('classification').agg({
    'total_volume_usd': 'sum',
    'total_pnl': 'sum'
}).reset_index()

# Step 3: Compute aggregated PnL per USD
pnl_efficiency_summary['pnl_per_usd'] = pnl_efficiency_summary['total_pnl'] / pnl_efficiency_summary['total_volume_usd']
```

```python
pnl_efficiency_summary
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.bar(pnl_efficiency_summary['classification'], pnl_efficiency_summary['pnl_per_usd'])
plt.title('PnL per USD Traded by Market Sentiment')
plt.ylabel('PnL per USD Traded')
plt.xlabel('Market Sentiment')
```

Extreme Greed markets offer the best return per unit capital.

```python
# Step 1: Compute std deviation of avg_pnl_per_trade per sentiment
pnl_volatility = merged_df.groupby('classification')['avg_pnl_per_trade'].std().reset_index()
pnl_volatility = pnl_volatility.rename(columns={'avg_pnl_per_trade': 'pnl_std'})

# Step 2: Compute mean avg_pnl_per_trade per sentiment (if not already)
pnl_mean = merged_df.groupby('classification')['avg_pnl_per_trade'].mean().reset_index()
pnl_mean = pnl_mean.rename(columns={'avg_pnl_per_trade': 'pnl_mean'})

# Step 3: Merge mean and std
risk_summary = pnl_mean.merge(pnl_volatility, on='classification')

# Step 4: Compute Risk-Adjusted PnL (Sharpe-like ratio)
risk_summary['risk_adjusted_pnl'] = risk_summary['pnl_mean'] / risk_summary['pnl_std']
```

```python
risk_summary
```

```python
plt.figure(figsize=(8,5))
plt.bar(risk_summary['classification'], risk_summary['risk_adjusted_pnl'])
plt.title('Risk-Adjusted PnL by Market Sentiment')
plt.ylabel('Risk-Adjusted PnL (Avg PnL / Std Dev)')
plt.xlabel('Market Sentiment')
```

1. Greed & Extreme Greed → highest risk-adjusted PnL (~0.146)

Trades are relatively profitable and moderately consistent, making them ideal for strategic trading.

2. Neutral → moderate risk-adjusted PnL (~0.136)

Lower than Greed but still reasonable stability.

3. Fear → lower (~0.114)

Despite high total PnL, high volatility reduces risk-adjusted performance.

4. Extreme Fear → extremely low (~0.007)

Lots of small trades with huge volatility → very risky per trade.

💡 Key Takeaways

Most profitable & stable trades: Greed and Extreme Greed markets

High volume but risky trades: Extreme Fear and Fear markets

Neutral markets are moderately profitable with decent stability

```python
# Step 1: Merge avg PnL per trade with PnL per USD
combined = sentiment_summary.merge(
    pnl_efficiency_summary[['classification','pnl_per_usd']],
    on='classification'
)

# Step 2: Merge with risk-adjusted PnL
combined = combined.merge(
    risk_summary[['classification','risk_adjusted_pnl']],
    on='classification'
)

# Step 3: Keep only relevant columns and rename for clarity
combined_summary = combined[['classification','avg_pnl_per_trade','pnl_per_usd','risk_adjusted_pnl']]
combined_summary.columns = ['Sentiment','Avg PnL per Trade (USD)','PnL per USD Traded','Risk-Adjusted PnL']
```

```python
combined_summary
```

```python
# Select metrics to plot
plot_data = combined_summary.set_index('Sentiment')[['Avg PnL per Trade (USD)','PnL per USD Traded','Risk-Adjusted PnL']]

# Scale PnL per USD and Risk-Adjusted PnL for visibility
plot_data['PnL per USD Traded'] *= 1000
plot_data['Risk-Adjusted PnL'] *= 1000

# Plot
plot_data.plot(kind='bar', figsize=(10,6))
plt.ylabel('Values (scaled for visibility)')
plt.title('Trader Performance Metrics by Market Sentiment')
plt.xticks(rotation=45)
plt.tight_layout()
```
